```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

```python
x = np.linspace(1 , 10 , 100)
x
```

```python
y = np.log(x)
y1 = np.log2(x)
y2 = np.log10(x)
```

```python
import seaborn as sns
sns.set()

sns.set_style('whitegrid' , {'grid.color' : 'green'})

plt.plot(x ,y)
plt.plot(x , y1)
plt.plot(x , y2)
```
