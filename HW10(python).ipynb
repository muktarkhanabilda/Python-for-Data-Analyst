```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0 , 10 , 100)

y = np.exp(x)

plt.plot(x , y)


```

```python
y = np.log(x)
y1 = np.log2(x)
y2 = np.log10(x)

plt.plot(x , y , '--g')
plt.plot(x , y1 , '-.')
plt.plot(x , y2 , ':r')
```

```python
x = np.linspace(0 , 1 , 100)

y = np.arcsin(x)
y1 = np.arccos(x)
y2 = np.arctan(x)

plt.plot(x , y , '-r')
plt.plot(x , y1 , '--g')
plt.plot(x , y2 , '-.y')

plt.savefig('trigonometric.png')
```

```python
x = np.linspace(0 , 10, 100)

y = np.sinh(x)

ax = plt.axes()

ax.plot(x , y, 'blue')

ax.set(xlim = (-2, 12), ylim = (-1000, 12000) , xlabel = 'Ленейный массив' , ylabel = 'sinh(x)' , title = 'График гиперболического синуса')
```

```python
y = np.log(x)
y1 = np.log2(x)
y2 = np.log10(x)
y3 = np.arcsin(x)
y4 = np.arccos(x)
y5 = np.arctan(x)

fig, axs = plt.subplots( 2 , 3)

axs[0 , 0].plot(x , y, 'blue')
axs[0 , 0].set_title('График log(x)')

axs[0 , 1].plot(x , y1, 'g')
axs[0 , 1].set_title('График log2(x)')

axs[0 , 2].plot(x , y2 , 'red')
axs[0 , 2].set_title('График log10(x)')

axs[1 , 0].plot(x , y3 , 'purple')
axs[1 , 0].set_title('График arcsin(x)')

axs[1, 1].plot(x , y4, 'orange')
axs[1 , 1].set_title('График arccos(x)')

axs[1 , 2].plot(x , y5, 'brown')
axs[1 , 2].set_title('График arctan(x)')

for ax in axs.flat:
    ax.legend()
```
